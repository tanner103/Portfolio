### Simple Linear Regression

#### 1. Build a function that takes in two lists, Y and X, and returns the intercept and slope coefficients that minimize SSE/RSS. 

```python
#Write linreg function.. 

def linreg(x,y):
    y_bar = y.mean()
    x_bar = x.mean()
    count = len(x)
    top = 0
    bottom = 0
    for i in range(count):
        top += (x[i] - x_bar) * (y[i] - y_bar)
        bottom += (x[i] - x_bar) ** 2
    beta_1 = top / bottom
    beta_0 = y_bar - (beta_1 * x_bar)
    print(f'The intercept is: {beta_0} The slope is: {beta_1}')
```

```python
#Run a LinearRegression to check my data.
X = df[['Temp2']]
y = df['Count']

lr = LinearRegression()
lr.fit(X,y)

print(f'The slope is: {lr.coef_}')
print(f'The intercept is: {lr.intercept_}')
```

```python
#Line of best fit
line = lr.intercept_ + (lr.coef_ * X)
```

```python
#Plot with line fo best fit
ax = plt.gca()
plt.xlabel("Temperature")
plt.ylabel("Count")
ax.scatter(X,y)
ax.plot(X, line, color='r');
```

#### 3. Build a function that takes in a list of predictions and a corresponding list of observed values and then prints the following metrics: mean squared error, root mean squared error, R^2.

```python
def regmetrics(y,predictions):
    print('MSE:', ((predictions - y)**2).mean())
    print('RMSE:', (((predictions - y)**2).mean()**.5))
    print('R^2:',(1 - (((y-predictions) **2).sum()) / (((y - np.mean(y))**2).sum())))
```
