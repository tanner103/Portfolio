```python
#Import needed Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob

%matplotlib inline
```

```python
#Import all stats data sets and combine
path = './data' # use your path
files = glob.glob(path + "/df_2*.csv")

li = []

for i in files:
    dataframes = pd.read_csv(i, index_col=None, header=0)
    li.append(dataframes)

df = pd.concat(li, axis=0, ignore_index=True).drop('Unnamed: 0',axis=1)
```

```python
df
```

```python
#Read in All-Star data
df_as = pd.read_csv('./data/df_as.csv', index_col=None, header=0)
df_as.head()
```

```python
#Shape
df_as.shape
```

```python
#Clean up column names
df_as.columns = ['PLAYER', 'All-Star']
```

```python
#Merge together my two data sets
df = df.merge(df_as,how='outer', on='PLAYER')
```

```python
#Remove all null values
df['All-Star'] = df['All-Star'].fillna(0)
```

```python
#Review new dataframe
df.head()
```

```python
#Review full dataframe
df.info()
```

```python
#Review full dataframe
df.describe()
```

```python
#Ceate a column for Year
df['YEAR'] = [int(i[-4:]) for i in df['PLAYER']]
```

```python
#Save new DataFrame
df.to_csv('./data/df.csv',index=False)
```
